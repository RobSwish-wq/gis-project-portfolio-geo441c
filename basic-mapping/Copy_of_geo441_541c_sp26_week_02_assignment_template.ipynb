## **Reflection on data management function in ArcGIS Pro**

- **Write a 100-word paragraph to describe the relationship between**
  - **`Geodatabase`**
  - **`Feature Dataset`**
  - **`Feature Classes`**
  - I can refer to **Page 9** of [assignment guide](https://miamioh.instructure.com/courses/246872/files/folder/week-02-resources?preview=38195670) or other resources online, such as
    - [Fundamentals of geodatabase](https://pro.arcgis.com/en/pro-app/latest/help/data/geodatabases/overview/fundamentals-of-the-geodatabase.htm)
    - [Feature datasets in ArcGIS Pro](https://pro.arcgis.com/en/pro-app/latest/help/data/feature-datasets/feature-datasets-in-arcgis-pro.htm)
- A geodatabase is a collection of various types of datasets that is stored for future use, such as small or large data tables, graphs, etc. used by multiple viewers. That way, it is easier to access something that requires the same data as, say, a previous project.
- The main source of information storage in geodatabases are what's called feature datasets. What these do is store things called feature classes, which are collections of similar or common data, like polygons, lines, and more.
- To put it simply, feature datasets are collections of feature classes, and geodatabases are collections of feature datasets.
